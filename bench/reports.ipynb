```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_json("./reports.jsonl", lines=True)
```

```python
df.fun_impl.unique()
```

```python
ax1 = df.where(df.fun_name == "take").where(df.fun_impl == "kit-ty-kate").plot(x="list_size", y="time_delta", label="kit-ty-kate")
df.where(df.fun_name == "take").where(df.fun_impl == "kit-ty-kate 2.").plot(x="list_size", y="time_delta", label="kit-ty-kate 2.", ax=ax1)
#df.where(df.fun_name == "take").where(df.fun_impl == "Containers").plot(x="list_size", y="time_delta", label="containers", ax=ax1)
#df.where(df.fun_name == "take").where(df.fun_impl == "Base").plot(x="list_size", y="time_delta", label="base", ax=ax1)
df.where(df.fun_impl == "take-custom-mod-cons").plot(x="list_size", y="time_delta", label="custom-mod-cons", ax=ax1)
ax1.set_ylabel("time delta in s")
```

```python
ax1 = df.where(df.fun_name == "take").where(df.fun_impl == "kit-ty-kate").plot(x="list_size", y="alloc_delta", label="kit-ty-kate")
df.where(df.fun_name == "take").where(df.fun_impl == "kit-ty-kate 2.").plot(x="list_size", y="alloc_delta", label="kit-ty-kate 2.", ax=ax1)
#df.where(df.fun_name == "take").where(df.fun_impl == "Containers").plot(x="list_size", y="alloc_delta", label="containers", ax=ax1)
#df.where(df.fun_name == "take").where(df.fun_impl == "Base").plot(x="list_size", y="alloc_delta", label="base", ax=ax1)
df.where(df.fun_impl == "take-custom-mod-cons").plot(x="list_size", y="alloc_delta", label="custom-mod-cons", ax=ax1)
ax1.set_ylabel("heap alloc delta")
```

```python
ax1 = df.where(df.fun_name == "drop").where(df.fun_impl == "kit-ty-kate").plot(x="list_size", y="time_delta", label="kit-ty-kate")
df.where(df.fun_name == "drop").where(df.fun_impl == "Containers").plot(x="list_size", y="time_delta", label="containers", ax=ax1)
df.where(df.fun_name == "drop").where(df.fun_impl == "Base").plot(x="list_size", y="time_delta", label="base", ax=ax1)
ax1.set_ylabel("time delta in s")
```

```python
ax1 = df.where(df.fun_name == "drop").where(df.fun_impl == "kit-ty-kate").plot(x="list_size", y="alloc_delta", label="kit-ty-kate")
df.where(df.fun_name == "drop").where(df.fun_impl == "Containers").plot(x="list_size", y="alloc_delta", label="containers", ax=ax1)
df.where(df.fun_name == "drop").where(df.fun_impl == "Base").plot(x="list_size", y="alloc_delta", label="base", ax=ax1)
#df.where(df.fun_impl == "take-custom-mod-cons").plot(x="list_size", y="alloc_delta", label="custom-mod-cons", ax=ax1)
ax1.set_ylabel("heap alloc delta")
```
